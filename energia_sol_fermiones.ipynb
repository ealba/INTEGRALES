```julia
using Cubature
using graf
using PyPlot
```

```julia
m = 6.0151223*1.660538782e-27;
hbarr = 1.054571628e-34;
Kb = 1.3806488e-23;
T = 0.0;
λ=1;


ωz = 3800*λ;
ω0=(ωz^(2/3))/λ^(1/3);
Npart = 1e3;
a0=1.02;


Ef=hbarr*(ωz/λ)*(6*λ*Npart)^(1/3);
Kf=sqrt(2*m*Ef)/hbarr;
Rf(a)=sqrt(2*Ef/(m*a^2*ω0^2));

z0c=100/Kf;
```

```julia
function ntf(r::Float64,a::Float64)
    if  r<Rf(a)
        return (8/π^2)*(Npart*λ/Rf(a)^3)*(1-(r/Rf(a))^2)^(3/2) 
    end
    return 0    
end


function Normntf(a::Float64)
    (16/35)*Npart*Rf(a)*((8*λ)/(π^2*Rf(a)^3))^2;
end

function ntfnorm(r::Float64,a::Float64)
    ntf(r,a)/sqrt(Normntf(a))
end

ntf(r)=ntf(r,a0)
ntfnorm(r)=ntfnorm(r,a0)
```

```julia
F(r)=conj(ntfnorm(r))*ntfnorm(r);
(val,err) = hquadrature(F, 0, Rf(a0); reltol=1e-8, abstol=0, maxevals=0)
```

### (*(del simple variational aproach para a=1.02)*) *No conocs la depedendia explícita con la longitud dispersión,   pero conoces la tabla II*)

```julia
function ΨsolTF(r::Float64, z0::Float64, a::Float64, γ::Float64) 
    sqrt(ntf(r, a))*(  tanh((r-z0)/(γ/Kf))/γ + im*sqrt(1 - 1/γ^2)  ); 
end


Norma_cuadrada_ΨsolTf(r,z0,a,γ)= (ntf(r, a)/Npart)  *  (1 - sech((r - z0)/(γ/Kf))^2/γ^2)

function NormΨsolTF(z0::Float64, a::Float64, γ::Float64) 
    f(r)=Norma_cuadrada_ΨsolTf(r,z0,a,γ)
    (val,err) = hquadrature(f, 0, Rf(a); reltol=1e-8, abstol=0, maxevals=0)
    return (val,err)
end

ΨsolTFnorm(r, z0, a, γ)=ΨsolTF(r, z0, a, γ)/sqrt(NormΨsolTF(z0, a, γ)[1]);
```

#### comprobar que esté bien normalizada ΨsolTFnorm

```julia
G(r)=norm(ΨsolTFnorm(r, 100/Kf, a0, 1.0))^2;
    (val,err) = hquadrature(G, 0, Rf(a0); reltol=1e-8, abstol=0, maxevals=0)
```

# LAS EXPRESIONES PARA LA ENERGÍA

```julia
function ecinedinamica(r::Float64,z0::Float64,a::Float64,γ::Float64)
    (hbarr^2/2m)*(ntf(r,a)/NormΨsolTF(z0,a,γ)[1])*(  (3/2)*(r/(r^2-Rf(a)^2))*tanh((r-z0)/(γ/Kf))/γ +(sech((r- z0)/(γ/Kf)))^2/(γ^2/Kf)  )^2
end

function ecineinercial(r::Float64,z0::Float64,a::Float64,γ::Float64)
    (hbarr^2/2m)*ntf(r,a)*((3/2)*(r/(r^2-Rf(a)^2)))^2 * ((1-γ^(-2.0))*Normntf(a)-NormΨsolTF(z0,a,γ)[1])/(Normntf(a)*NormΨsolTF(z0,a,γ)[1])

end

ecinetotal(r,z0,a,γ)=ecinedinamica(r,z0,a,γ)+ecineinercial(r,z0,a,γ);


function epottotal(r::Float64,z0::Float64,a::Float64,γ::Float64)
    (m*ω0^2*r^2*ntfnorm(r,a)/2)*((1-(sech((r-z0)/(γ/Kf)))^2/γ^2))*(1/NormΨsolTF(z0,a,γ)[1]-1/Normntf(a))
end

function eprestotal(r::Float64,z0::Float64,a::Float64,γ::Float64)
    (-g*ntf(r,a)^2/2)*(     (  1- (sech((r-z0)/(γ/Kf)))^2/γ^2  )^2 *( NormΨsolTF(z0,a,γ)[1]^(-2.0)-Normntf(a)^(-2.0) )     )
end

esoliton(r,z0,a,γ)=ecinetotal(r,z0,a,γ)+epottotal(r,z0,a,γ)+eprestotal(r,z0,a,γ);
```

# Integrar la densidad de energía del solitón

```julia
function Esoliton(N,z0::Float64,a::Float64,γ::Float64)
    integrando(r)=esoliton(r,z0,a,γ);
    minimo_integracion=z0-N/Kf;
    maximo_integracion=z0+N/Kf;
    (val,err) = hquadrature(integrando, minimo_integracion, maximo_integracion; reltol=1e-8, abstol=0, maxevals=0)    
    return val
end
```

```julia
function Energia_soliton_entre_Ef(N,z0::Float64,a::Float64,γ::Float64)
    Esoliton(N,z0,a,γ)/Ef;
end

######Parámetros estándar
N=10;
a=a0;
z0=z0c;
γ=1.0;
```
