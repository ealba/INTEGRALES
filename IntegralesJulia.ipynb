```julia
function U(v::Vector{Float64})
                    (1/2)*m*(ωr*v[1])^2 + ωz*(v[2]-z0)^2
                    end
```

```julia
function TF0(v::Vector{Float64})
                    if  μ-U(v)> 0
                        return sqrt((μ-U(v))/g_α)*(tanh((v[2]-z0)/ (ξ*γ)) + β*im)
                    end
                    return 0
                    end
```

```julia
function BESSEL(v::Vector{Float64})
                    besseli(0, -2*r*v[1]*(A/sin(ωr*t))) 
                    end
    
```

```julia
function Ψreal(v::Vector{Float64})
    # rho, z, t ctes
    real(-(2*π*A/sin(ωr*t))  * (B/sqrt(sin(ωz*t))) *exp((A/sin(ωr*t))*(r^2+v[1]^2)*cos(ωr*t)) *exp((B/sqrt(sin(ωz*t)))*(z^2+v[2]^2)*cos(ωz*t)-2*z*v[2])*TF0(v)*BESSEL(v)) 
    end
```

```julia
function Ψimag(v::Vector{Float64})
    # rho, z, t ctes
    imag(-(2*π*A/sin(ωr*t))  * (B/sqrt(sin(ωz*t))) *exp((A/sin(ωr*t))*(r^2+v[1]^2)*cos(ωr*t)) *exp((B/sqrt(sin(ωz*t)))*(z^2+v[2]^2)*cos(ωz*t)-2*z*v[2])*TF0(v)*BESSEL(v)) 
    end
```
